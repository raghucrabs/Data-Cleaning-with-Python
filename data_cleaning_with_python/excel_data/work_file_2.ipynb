## Import necessary libraries


```python
import pandas as pd
```

## Reading the Excel File into a pandas dataframe


```python
df = pd.read_excel("./raw_data/GDPpercapita.xlsx",
                  sheet_name = "OECD.Stat export",
                  skiprows = 4,
                  skipfooter = 1,
                  usecols = "A, C:T")
```

## Data Cleaning

```python
# Renaming the name of the year column

df.rename(columns = {'Year':'metro'}, inplace = True)

df.head()
```

```python
# Print unique datatypes in the column

print(df.metro.map(type).unique())
```

```python
# Count Non-String values in a column

count_non_strings = df.metro.apply(lambda x: not isinstance(x,str)).sum()

print(f"Non-string values in 'metro': {count_non_strings}")
```

```python
# Convert non-string values to strings in 'metro' column

df.metro = df.metro.astype(str)
```

```python
# Check for the datatype in 'metro' column

df["metro"].map(type).unique()
```

```python
# After converting to string datatype, we can check for leading 
# trailing spaces 

df.metro.str.startswith(' ').any()
```

```python
# Check for trailing spaces

df.metro.str.endswith(' ').any()
```

```python
# Remove trailing and leading spaces

df.metro = df.metro.str.strip()
```

```python
# Convert the data columns to numeric and rename the column names

for col in df.columns[1:]:
    df[col] = pd.to_numeric(df[col],errors = "coerce")#convert to numeric 
    df.rename(columns = {col : 'perGDP'+col},inplace = True)

                            
```
